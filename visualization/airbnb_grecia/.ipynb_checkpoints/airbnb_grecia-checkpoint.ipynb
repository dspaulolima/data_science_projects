### Sobre o Airbnb

O Airbnb é uma empresa que opera como um serviço online de aluguél de casas, apartamentos ou quartos por agendamento durante todo o calendário. Está presente em todos os cantos do mundo e funciona oferecendo imóveis de terceiros, cadastrados no serviço, para aluguél. Dessa forma age como um intermediário entre o anfitrião e o inquilino, ganhando uma comissão por cada reserva.

Foi fundada em 2008, na cidade de São Francisco - Califórnia, por Brian Chesky, Nathan Blecharczyk e Joe Gebbia. De acordo com o próprio site, o Airbnb é um dos maiores mercados do mundo para encontrar lugares únicos e autênticos para se hospedar e fazer várias coisas. Oferecendo mais de 7 milhões de acomodações e 50.000 atividades artesanais, todas alimentadas por anfitriões locais. Com um mecanismo de empoderamento econômico, o Airbnb ajudou milhões de empreendedores de hotelaria a monetizar seus espaços e paixões, mantendo os benefícios financeiros do turismo em suas próprias comunidades. Com mais de 750 milhões de hóspedes até o momento e acessível em 62 idiomas em mais de 220 países e regiões, o Airbnb promove conexão pessoa a pessoa, comunidade e confiança em todo o mundo.

### Sobre a Grécia
*Fontes*: [Eurodicas](https://www.eurodicas.com.br/tudo-sobre-a-grecia/) e [Wikipédia](https://pt.wikipedia.org/wiki/Gr%C3%A9cia)

A Grécia é um país do sudeste da Europa com milhares de ilhas espalhadas pelos mares Egeu e Jônico. Bastante influente na antiguidade, a nação é considerada o berço da civilização ocidental. Atenas, sua capital, conserva monumentos como a Acrópole, do século V a.C., onde fica o templo Partenon. A Grécia também é conhecida por suas praias, como Santorini, com suas areias escuras, e os festivos complexos hoteleiros de Míconos.

A Grécia tem um território de 131.957 km². Sua área é um pouco menor do que o Ceará, no Brasil. A cidade possui clima mediterrâneo e suas principais atividades estão voltadas para o turismo, o comércio e serviços. Cerca de 96% da população são gregos e o idioma oficial da Grécia é o grego. A moeda oficial da Grécia é o euro. 

**Outros fatos importantes na história grega**

* Em 776 a.C. foram realizados os Primeiros Jogos Olímpicos da história, na cidade grega de Olímpia;
* Em 594 a.C., Sólon, legislador grego, iniciou uma grande reforma política, econômica e social em Atenas;
* Em 490 a.C, os gregos venceram os persas na batalha de Maratona;
* Em 478 a.C, Atenas implementou a Liga de Delos para defender as cidades gregas contra-ataques persas, nas Guerras Médicas.

## Importação dos pacotes e carga dos dados

**Q2 - QUAL A QUANTIDADE DE IMÓVEIS LISTADOS POR CIDADE?**

```python
airbnb['neighbourhood'].value_counts()
```

Nossa! Muitas cidades. Como o grupo gosta de locais mais badalados, decidimos pesquisar somente as 10 cidades com maior quantidade de imóveis listados.

```python
### Quantidade de hospedagem por cidade

# Coleta de valores e nomes das cidades
neighbourhood_values = airbnb['neighbourhood'].value_counts().sort_values(ascending=False)[0:10]
neighbourhood_names = neighbourhood_values.keys().unique()

# Plotagem do gráfico
sns.barplot(data=airbnb, x=neighbourhood_values, y=neighbourhood_names)
plt.title('Listings x Neighbourhood', size=13)
plt.xlabel('Total of listings', size=12)
plt.ylabel('Neighbourhood', size=12)
plt.show()
```

OK! A primeira cidade, ΕΜΠΟΡΙΚΟ ΤΡΙΓΩΝΟ-ΠΛΑΚΑ e ΚΟΥΚΑΚΙ-ΜΑΚΡΥΓΙΑΝΝΗ, é de fato um lugar em que o Airbnb atua com força (passando de mil listagens)! Como gostamos de locais com bastante movimento, vamos pegar as 10 maiores em quantidade de imóveis listados.

```python
# TOP10 - Filtrando as cidades e dispondo em um novo dataframe
neighboors = airbnb['neighbourhood']

top10 = airbnb[
    (neighboors == 'ΕΜΠΟΡΙΚΟ ΤΡΙΓΩΝΟ-ΠΛΑΚΑ') |
    (neighboors == 'ΚΟΥΚΑΚΙ-ΜΑΚΡΥΓΙΑΝΝΗ') |
    (neighboors == 'ΜΟΥΣΕΙΟ-ΕΞΑΡΧΕΙΑ-ΝΕΑΠΟΛΗ') |
    (neighboors == 'ΝΕΟΣ ΚΟΣΜΟΣ') |
    (neighboors == 'ΑΓΙΟΣ ΚΩΝΣΤΑΝΤΙΝΟΣ-ΠΛΑΤΕΙΑ ΒΑΘΗΣ') |
    (neighboors == 'ΠΑΓΚΡΑΤΙ') |
    (neighboors == 'ΠΛΑΤΕΙΑ ΑΤΤΙΚΗΣ') |
    (neighboors == 'ΚΟΛΩΝΑΚΙ') |
    (neighboors == 'ΑΚΡΟΠΟΛΗ') |
    (neighboors == 'ΑΜΠΕΛΟΚΗΠΟΙ')]
```

**Q3 - DO TOP 10, QUAL A DISTRIBUIÇÃO DOS TIPOS DE HOSPEDAGEM?**

```python
### Plotagem da quantidade dos tipos de hospedagem

# Nomes e valores dos tipos de hospedagem
room_type_names = top10['room_type'].unique()
room_type_values = top10['room_type'].value_counts()

sns.barplot(data=top10, x=room_type_names, y=room_type_values)
plt.title('Room Types of Top 10 Neighbours', size=13)
# plt.xlabel('Room type', size=12)
plt.ylabel('Counting', size=12)
plt.show()

sns.catplot(data=top10, x='neighbourhood', hue='room_type', kind="count",
            edgecolor=".6", aspect=1.4, legend_out=False);
plt.xticks(rotation=45)
plt.title('Room Types Counting x Neighbourhood', fontsize=13)
# plt.xlabel('Neighbourhood', fontsize=12)
plt.ylabel('Counting', fontsize=12)

plt.show()

# Contagem da quantidade por tipo de hospedagem
top10['room_type'].value_counts()
```

Interessante! Mas como suspeitado. Já era fato de que, como grupo, queríamos locar um imóvel inteiro. E percebemos que a predominância é do tipo que procurávamos, inclusive quando olhamos por cidade.

**Q4 - COMO ANDAM OS PREÇOS??**

```python
# Agrupamento por tipo de quarto, retornando preços médios.
price_mean_rooms = top10.groupby('room_type')['price'].agg('mean')

# Plotagem
price_mean_rooms.plot(kind='barh')
plt.title('Mean Price of Room Types', size=13)
plt.xlabel('Mean Price', size=12)
plt.ylabel('Room Type', size=12)

plt.annotate(
    round(price_mean_rooms['Shared room'],2),  # label escrita
    xy=((round(price_mean_rooms['Shared room'],2)/4), 2.90), ha='center', color='w', fontsize=14)
plt.annotate(
    round(price_mean_rooms['Private room'].mean(),2),  # label escrita
    xy=((round(price_mean_rooms['Shared room'],2)/4), 1.90), ha='center', color='w', fontsize=14)
plt.annotate(
    round(price_mean_rooms['Hotel room'],2),  # label escrita
    xy=((round(price_mean_rooms['Shared room'],2)/4), 0.90), ha='center', color='w', fontsize=14)
plt.annotate(
    round(price_mean_rooms['Entire home/apt'],2),  # label escrita
    xy=((round(price_mean_rooms['Shared room'],2)/4), -0.1), ha='center', color='w', fontsize=14)

plt.show()
```

Opa! A média de preço em Hotel é bem acima do restante. Aqui atestamos que é mais conveniente (e econômico) ir para um imóvel inteiro.

```python
# plt.figure(figsize=(12,8))
sns.scatterplot(data=top10, x='neighbourhood', y='price', hue='room_type')
plt.xticks(rotation=45)
plt.axhline(top10.price.mean(), c='orange')
plt.title('Price x Neighbourhood with Yellow Mean Price Line', fontsize=13)
plt.xlabel('Neighbourhood', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.show()
```

Uau! Definitivamente não vamos ficar naquele *outlier* da cidade de ΚΟΥΚΑΚΙ-ΜΑΚΡΥΓΙΑΝΝΗ. 7 mil euros? Estamos fora dessa, obrigado! Por falar nisso, e por curiosidade, qual seria o status para cada imóvel? Vamos definí-los como ``low_cost``, ``medium_cost``, ``high_cost`` e ``luxe_cost``.

1. Olhar os valores mínimo, máximo, média e os quartis para definir o status de cada preço; e
2. Transformar em REAL, para comparar com o que consideramos razoável pagar. *(1 EURO = 5,77 REAIS)*.

```python
# Olhando as estatísticas dos preço em EURO
EURO = top10.describe()['price']

# Criando dicionário com valores em EURO e em REAL
EURO_REAL_dict = {
    'Euro': [EURO['mean'], EURO['min'], EURO['25%'], EURO['50%'], EURO['75%'], EURO['max']],
    'Real': [EURO['mean']*5.77, EURO['min']*5.77, EURO['25%']*5.77, EURO['50%']*5.77, EURO['75%']*5.77, EURO['max']*5.77]
            }

# Dataframe para facilitar a comparação
pd.DataFrame(EURO_REAL_dict, index=['mean', 'min', '25%', '50%', '75%', 'max']).round(2)
```

Os preços estão no padrão quando se pensa em economizar e locar em grupo *(salvo aquele imóvel de 40.390,00 reais)*. Para definir os preços, e levando em consideração o quanto queremos gastar com turismo, alimentação e presentes, consideramos o seguinte para pagar de hospedagem:

1. Até 207.72 reais (36 euros ) como ``low_cost``;
2. Acima de 207 até 288.50 reais (50 euros) como ``medium_cost``;
3. Acima de 288.50 até 461.60 reais (80 euros) como ``high_cost``; e
4. Acima de 461.60 como ``luxe_cost``

*Obs.: Existe UM imóvel com preço ZERO (por qualquer motivo). Decidimos mantê-lo pois não irá interferir na análise.*

```python
# Criando atributo com status de acordo com um range de preços
top10['price_status'] = ['low_cost' if (i > 0) & (i <= 36) else
                         'medium_cost' if (i > 36) & (i <= 50) else
                         'high_cost' if (i > 50) & (i <= 80) else
                         'luxe_cost' for i in top10['price']]
```

```python
### Gráfico com preço por status

# Médias individuais
low_count = top10['price_status'].value_counts()['low_cost']
medium_count = top10['price_status'].value_counts()['medium_cost']
high_count = top10['price_status'].value_counts()['high_cost']
luxe_count = top10['price_status'].value_counts()['luxe_cost']

# Gráfico
top10.price_status.value_counts().plot(kind='barh')
plt.title('Price Status Counting', fontsize=13)
plt.xlabel('Counting', fontsize=12)
plt.ylabel('Price Status', fontsize=12)

plt.annotate(
    round(luxe_count,2),  # label escrita
    xy=((round(luxe_count,2)/10), 2.90), ha='center', color='w', fontsize=14)
plt.annotate(
    round(medium_count,2),  # label escrita
    xy=((round(medium_count,2)/10), 1.90), ha='center', color='w', fontsize=14)
plt.annotate(
    round(low_count,2),  # label escrita
    xy=((round(low_count,2)/10), 0.90), ha='center', color='w', fontsize=14)
plt.annotate(
    round(high_count,2),  # label escrita
    xy=((round(high_count,2)/10), -0.1), ha='center', color='w', fontsize=14)

plt.show()
```

```python
top10.groupby('price_status')['price'].mean().plot.bar()
plt.title('Mean Price of Price Status', fontsize=13)
plt.xlabel('Price Status', fontsize=12)
plt.ylabel('Mean', fontsize=12)
plt.xticks(rotation=45)
plt.show()

top10.groupby('price_status')['price'].mean().round(2)
```

Existe um equilíbrio na quantidade de imóveis de acordo com o status do preço. Porém a média dos preços nos leva a escolha do ``medium_cost``. Temos, portanto, 1392 possibilidades de escolha.

**Q5 - QUE TAL DEFINIR MELHOR O QUE ESTAMOS PROCURANDO?**

* **Q5.1 - Queremos ficar quantas noites?** 30 noites. Vamos descartar o ``minimum_nights`` acima disso.
* **Q5.2 - Qual status de preço e tipo de hospedagem escolhemos?** ``medium_cost`` e ``Entire home/apt``
* **Q5.3 - Quais os imóveis com diponibilidade de, no mínimo, 30 dias?** Vamos selecionar ``avaliability`` a partir de 30 dias.
* **Q5.4 - Qual o número mínimo de reviews que queremos observar?** Escolhemos remover de ``number_of_reviews`` apenas os que não tem reviews.

```python
# Aplicando o filtro acima, ficamos com 1041 possibilidades
filtered_airbnb = top10[(top10['minimum_nights'] <= 30) & 
                       (top10['price_status'] == 'medium_cost') &
                       (top10['room_type'] == 'Entire home/apt') & 
                       (top10['availability_365'] >= 30) &
                       (top10['number_of_reviews'] > 0)]

print(f'We have found {filtered_airbnb.pipe(len)} listing available to choose.')
```

*Obs: Se você estiver visualizando no github, recomendo clicar no botão do Google Colab no início da página pois este é um mapa interativo. Salvo contrário, imprimiremos uma imagem salva com a mesma visualização.* 

Por questões de melhor visualização dos nomes gregos, resolvemos traduzi-los *(criando uma nova coluna)*. Utilizamos o Google Translator e o Google *(este quando aquele não conseguiu traduzir)*.

```python
## Novos atributos para auxílio na plotagem geográfica

# Cores
filtered_airbnb['colors'] = filtered_airbnb['neighbourhood'].apply(lambda x: 'lightblue' if x == 'ΑΜΠΕΛΟΚΗΠΟΙ' else
                                                                          'lightgreen' if x == 'ΕΜΠΟΡΙΚΟ ΤΡΙΓΩΝΟ-ΠΛΑΚΑ' else
                                                                          'orange' if x == 'ΠΑΓΚΡΑΤΙ' else
                                                                          'red' if x == 'ΠΛΑΤΕΙΑ ΑΤΤΙΚΗΣ' else
                                                                   'yellow' if x == 'ΑΓΙΟΣ ΚΩΝΣΤΑΝΤΙΝΟΣ-ΠΛΑΤΕΙΑ ΒΑΘΗΣ' else
                                                                   'black' if x == 'ΚΟΥΚΑΚΙ-ΜΑΚΡΥΓΙΑΝΝΗ' else
                                                                   'blue' if x == 'ΜΟΥΣΕΙΟ-ΕΞΑΡΧΕΙΑ-ΝΕΑΠΟΛΗ' else
                                                                   'gray' if x == 'ΚΟΛΩΝΑΚΙ' else
                                                                   'green' if x == 'ΝΕΟΣ ΚΟΣΜΟΣ' else
                                                                   'magenta')

# Traduções
translations = {'ΑΜΠΕΛΟΚΗΠΟΙ':'AMPELOKIPOI', 'ΕΜΠΟΡΙΚΟ ΤΡΙΓΩΝΟ-ΠΛΑΚΑ':'TRADE TRIANGLE-SIDE', 'ΠΑΓΚΡΑΤΙ':'PANGRATI',
       'ΠΛΑΤΕΙΑ ΑΤΤΙΚΗΣ': 'ATTICA SQUARE', 'ΑΓΙΟΣ ΚΩΝΣΤΑΝΤΙΝΟΣ-ΠΛΑΤΕΙΑ ΒΑΘΗΣ': 'AGIOS KONSTANTINOS-VATHIS SQUARE',
       'ΚΟΥΚΑΚΙ-ΜΑΚΡΥΓΙΑΝΝΗ':'KOUKAKI-MAKRYGIANNI', 'ΜΟΥΣΕΙΟ-ΕΞΑΡΧΕΙΑ-ΝΕΑΠΟΛΗ':'MUSEUM-EXARCHIA-NEAPOLI', 'ΚΟΛΩΝΑΚΙ':'KOLONAKI',
       'ΝΕΟΣ ΚΟΣΜΟΣ':'NEW WORLD', 'ΑΚΡΟΠΟΛΗ':'ACROPOLIS'}

filtered_airbnb['neighbourhood_translated'] = filtered_airbnb['neighbourhood'].map(translations)
```

```python
# Configuração do mapa
geomap = folium.Map(location=[37.97680, 23.72924], zoom_start = 14, tiles = "cartodbpositron")

for lat, lon, price, status, city, colors in zip(filtered_airbnb['latitude'], filtered_airbnb['longitude'],
                                              filtered_airbnb['price'],filtered_airbnb['price_status'], 
                                                 filtered_airbnb['neighbourhood_translated'], filtered_airbnb['colors']):
    folium.CircleMarker(
        [lat,lon],
        radius=2,
        popup= (
            'City:' + str(city).capitalize() + '<br>'
            'Price:' + str(price) + '<br>'
            'Status:' + str(status)
               ),
        color=colors,
        key_on=city,
        fill=True,
        fill_opacity=0.7        
    ).add_to(geomap)

# Plotagem
geomap
```
